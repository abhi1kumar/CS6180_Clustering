## Numpy

Package for doing linear algebra and operations on lots of data.

We can make vectors:

````python
import numpy as np

a = np.array([1,2,3])
b = np.array([0.0,1.0,0.0])
print 'shape of a:', a.shape, 'type of a:', a.dtype
print 'shape of b:', b.shape, 'type of b:', b.dtype
print 'dot product of a and b:', a.dot(b)
print 'elements of a added to elements of b:', a+b
````

We can also make matrices:

````python
A = np.array([[0,1],[1,0]])
B = np.array([[1,2,3,4]])
print 'shape of A:', A.shape
print 'shape of B:', B.shape

# Let's reshape B:
B = B.reshape((2,2))
print 'shape of B:', B.shape

# Now let's multiply them:
print 'Matrix multiply of A and B:'
print A.dot(B)
print 'Matrix multiply of B and A:'
print B.dot(A)
print 'Element-wise multiply of A and B:'
print A*B
````

## Scikit-learn

Scikit-learn is a machine learning library with many functions. We'll be using it heavily in the class.

````python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

random_state = 20160823

# Load data from file
df = pd.read_csv('blobs.txt')
X = df[['X0','X1']].get_values()
````

````python
# Incorrect number of clusters
df['ypred'] = KMeans(n_clusters=2, random_state=random_state).fit_predict(X)

# Let Seaborn pick the colors for us based on the hue
g = sns.FacetGrid(data=df, hue='ypred', size=10, subplot_kws=dict(aspect='equal')) 
g.map(plt.scatter,'X0','X1')
````

````python
# Correct number of clusters
df['ypred'] = KMeans(n_clusters=3, random_state=random_state).fit_predict(X)

# Let Seaborn pick the colors for us based on the hue
g = sns.FacetGrid(data=df, hue='ypred', size=10, subplot_kws=dict(aspect='equal')) 
g.map(plt.scatter,'X0','X1')
````
